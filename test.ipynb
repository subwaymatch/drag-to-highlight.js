```python
from bs4 import BeautifulSoup
import os
import re
import requests
import sys
import json
from pprint import pprint

news_url = 'http://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=101&oid=025&aid=0002728859'
```

```python
def get_basic_page(URL):
    url = URL
    return get_soup(url)

def get_soup(url):
    r = requests.get(url).text
    return BeautifulSoup(r, 'lxml')

def parse_basic_page(page):
    news={}
    news['date'] = get_date(page)
    news['contents'] = get_text(page)
    return news
        
def get_text(page):
    text = ''
    for item in page.find_all('div', id='articleBodyContents'):
        text = text + str(item.find_all(text=True))
    return text

def clean_text(text):
    cleaned_text = re.sub('[a-zA-Z]', '', text)
    cleaned_text = re.sub('[\{\}\[\]\/?,;:|\)*~`!^\-_+<>@\#$%&\\\=\(\'\"]','', cleaned_text)
    return cleaned_text

def get_date(page):
    return page.select('div[class=sponsor] span[class=t11]')[0].text
```

```python
news = parse_basic_page(get_basic_page(news_url))
```

```python
filename= "%d.json" % 0
print(filename)

print(news['contents'])

with open('../tmp/'+filename, 'w', ensure_ascii=False) as f:
    json.dump(news, f)
```
